## **Regresión Logística**
La regresión logística es un modelo estadístico utilizado para predecir la probabilidad de ocurrencia de un evento binario.  
- Se basa en la función sigmoide para mapear cualquier valor real a un rango entre 0 y 1.  
- Permite interpretar la influencia de cada variable independiente sobre la probabilidad del evento.

---

## **Máquinas de Vectores de Soporte (SVM)**
SVM es un algoritmo de aprendizaje supervisado que busca encontrar el **hiperplano óptimo** que separa las clases de manera máxima.  
- Puede manejar problemas lineales y no lineales mediante el uso de **kernels**.  
- Es robusto frente a alta dimensionalidad y es efectivo en conjuntos de datos pequeños y medianos.

---

## **Redes Neuronales (Multi-layer Perceptron)**
MLP es un tipo de red neuronal de alimentación hacia adelante que consiste en **capas de neuronas conectadas**.  
- Cada neurona aplica una función de activación no lineal, permitiendo modelar relaciones complejas.  
- Es adecuado para clasificación binaria y multiclase.

---

## **Kernels**
Los kernels permiten a SVM y otros modelos lineales **resolver problemas no lineales** proyectando los datos a un espacio de mayor dimensión.  
- Tipos comunes: lineal, polinomial, RBF (Radial Basis Function).  
- Permiten separar datos no linealmente separables sin necesidad de calcular explícitamente las transformaciones de alta dimensión.

---

## **Métricas para Clasificación**
Las métricas permiten evaluar el desempeño de un modelo:  
- **Accuracy**: porcentaje de predicciones correctas.  
- **Precision**: proporción de verdaderos positivos sobre todos los positivos predichos.  
- **Recall (Sensibilidad)**: proporción de verdaderos positivos sobre todos los positivos reales.  
- **F1-Score**: promedio armónico de precision y recall.  
- **ROC-AUC**: área bajo la curva ROC, útil para medir desempeño general en clasificación binaria.

---

## **Hiperparámetros**
Son parámetros del modelo que **no se aprenden durante el entrenamiento**, sino que se establecen antes:  
- Ejemplos en SVM: `C`, tipo de kernel, `gamma`.  
- Ejemplos en MLP: número de capas, número de neuronas por capa, tasa de aprendizaje.  
- Su correcta selección puede mejorar significativamente el rendimiento del modelo.

---

## **Optimización Bayesiana**
La optimización bayesiana es un enfoque probabilístico para **encontrar los mejores hiperparámetros** de un modelo de manera eficiente.  
- Construye un modelo probabilístico de la función objetivo y selecciona nuevos puntos basados en una función de adquisición.  
- Es más eficiente que la búsqueda exhaustiva (grid search) o aleatoria (random search).

# Análisis del Dataset: *Drug Consumption (Quantified)*


## **¿De dónde viene?**
> El dataset proviene del estudio *“The Five Factor Model of Personality and Evaluation of Drug Consumption Risk”* realizado por Fehrman et al. (2015) y publicado en *Data Science*.  
>  
> Fue donado el 16 de octubre de 2016 y está disponible públicamente en el repositorio de la *UCI Machine Learning Repository* bajo el nombre **Drug Consumption (Quantified)**.

---

## **¿Qué contiene?**
> Contiene información sobre **1885 individuos** y sus características demográficas, rasgos de personalidad medidos con el **modelo de los cinco grandes (Big Five)**, junto con indicadores de **consumo de distintas drogas** (legales e ilegales).  
>  
> En total, el dataset tiene **32 columnas**:  
> - 12 variables predictoras numéricas (edad, género, educación, país, etnicidad, rasgos de personalidad, impulsividad y búsqueda de sensaciones).  
> - 19 variables de salida que indican el nivel de consumo de diferentes sustancias.  
>  
> Para este proyecto, se utiliza una de ellas como variable objetivo —por ejemplo, **Cannabis**— transformada en una variable binaria (0 = no usuario, 1 = usuario).

---

## **¿Qué información dan las muestras?**
> Cada fila representa un participante anónimo con sus características psicológicas y sociodemográficas, y su historial de consumo de drogas.  
>  
> Las variables numéricas codifican la edad, nivel educativo, país de residencia, etnicidad y puntuaciones estandarizadas en diferentes rasgos de personalidad.  
>  
> Las muestras permiten explorar **patrones entre rasgos psicológicos y la propensión al consumo de sustancias**.

---

## **¿Qué se quiere analizar?**
> Se busca **predecir la probabilidad de consumo de una sustancia específica (por ejemplo, Cannabis)** a partir de las variables psicológicas y demográficas.  
>  
> En términos de modelado, se compararán distintos **modelos de clasificación binaria**:  
> - Regresión logística con regularización  
> - Máquina de vectores de soporte (SVM) con kernel RBF  
> - Perceptrón multicapa (MLP)  
>  
> Cada modelo será ajustado con **optimización de hiperparámetros mediante procesos gaussianos**, y evaluado usando **k-fold cross-validation**, reportando **media y desviación estándar** del desempeño.

---

## **¿Qué variables se tienen que transformar para poder usarse en un modelo de regresión?**
>Se selecciono una sustancia objetivo para predecir si el usaurio consume o no consume, se tansformo de manera binaria

---

## **¿Qué transformaciones se van a usar?** 
> - **Creación de variable binaria de salida** (por ejemplo, `Cannabis = 1` si ha consumido en el último año, mes o semana; `0` si nunca o hace más de una década).  
> - **División de datos** en conjuntos de entrenamiento y prueba, aplicando transformaciones dentro de un **pipeline de preprocesamiento**.

---

## **¿Qué resultado se podría encontrar al realizar una regresión?**
> Se espera encontrar **relaciones significativas entre ciertos rasgos de personalidad y la probabilidad de consumo**.  
>   
> Además, se obtendrán métricas de desempeño como **ROC-AUC** que permitirán comparar los modelos y determinar cuál predice mejor el consumo de Cannabis u otra droga seleccionada.

---



```python
import pandas as pd
import warnings

warnings.filterwarnings("ignore")


# Cargar el archivo tipo data
df = pd.read_csv("drug_consumption.data", header=None)

# Asignar nombre a las columnas
columns = [
    "ID", "Age", "Gender", "Education", "Country", "Ethnicity",
    "Nscore", "Escore", "Oscore", "Ascore", "Cscore", "Impulsive", "SS",
    "Alcohol", "Amphetamines", "Amyl_nitrite", "Benzos", "Caffeine",
    "Cannabis", "Chocolate", "Cocaine", "Crack", "Ecstasy",
    "Heroin", "Ketamine", "Legal_highs", "LSD", "Methadone",
    "Mushrooms", "Nicotine", "Semeron", "VSA"
]
df.columns = columns

# Poner la columna ID como índice
df = df.set_index("ID")

# Quitar consumo de drogas que no se analizarán
df = df.drop(columns =["Alcohol", "Amphetamines", "Amyl_nitrite", "Benzos", "Caffeine",
            "Chocolate", "Cocaine", "Crack", "Ecstasy", "Heroin", "Ketamine",
            "Legal_highs", "LSD", "Methadone", "Mushrooms", "Nicotine", "VSA", "Semeron"],
                        axis = 1)

# Crear variable binaria para analizar droga elegida: 1 = usuario, 0 = no usuario
df["Cannabis"] = df["Cannabis"].apply(lambda x: 0 if x in ["CL0", "CL1"] else 1)


df.head()
```

<div style="background-color: #d4edda; padding: 15px; border-radius: 8px;">
  <h3> Modelo 1 - Regresión logística con regularización</h3>
</div>


```python
import pandas as pd
import numpy as np
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, make_scorer
from skopt import BayesSearchCV
from skopt.space import Real, Categorical
```

```python
# Definir columnas
num_columns = [
    'Age', 'Gender', 'Education', 'Country', 'Ethnicity',
    'Nscore', 'Escore', 'Oscore', 'Ascore', 'Cscore',
    'Impulsive', 'SS'
]

cat_columns = []

target = ['Cannabis']

# Separar datos
X = df[num_columns + cat_columns]
y = df[target].values.ravel()
```

```python
# Definir pipeline base
pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('model', LogisticRegression(max_iter=1000))
])

# Espacio de búsqueda bayesiana (GP)
search_spaces = {
    'model__C': Real(1e-3, 100, prior='log-uniform'),   
    'model__penalty': Categorical(['l1', 'l2']),        
    'model__solver': Categorical(['liblinear', 'saga'])
}

# Definir validación cruzada y métrica AUC
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
auc_scorer = make_scorer(roc_auc_score, needs_threshold=True)

# Optimización Bayesiana (Procesos Gaussianos)
opt = BayesSearchCV(
    estimator=pipe,
    search_spaces=search_spaces,
    n_iter=30,              # número de combinaciones que probará
    scoring=auc_scorer,
    cv=kfold,               # K-Folds = 10
    n_jobs=-1,
    random_state=42,
    verbose=0
)

opt.fit(X, y)

# Resultados del CV (cross-validation)
cv_results = opt.cv_results_

# Extraer la mejor configuración, AUC promedio y desviación estándar
best_index = opt.best_index_
mean_auc = cv_results['mean_test_score'][best_index]
std_auc = cv_results['std_test_score'][best_index]

# Resultados
print("\nMejor configuración encontrada:")
print(opt.best_params_)

print(f"\nMejor AUC: {opt.best_score_:.4f}")
print(f"AUC promedio (10-Fold): {mean_auc:.4f}")
print(f"Desviación estándar AUC: {std_auc:.4f}")
```

<div style="background-color: #e6ccff; padding: 15px; border-radius: 8px;">
  <h3> Modelo 2 - Máquina de vectores de soporte con kernel RBF</h3>
</div>


```python
import numpy as np
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import roc_auc_score, RocCurveDisplay
import matplotlib.pyplot as plt
from skopt import BayesSearchCV
from skopt.space import Real, Categorical

X = df.drop(columns=["Cannabis"])
y = df["Cannabis"].astype(int)

Xtrain, Xtest, ytrain, ytest = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("svc", SVC(kernel="rbf", probability=True, random_state=42))
])

search_spaces = {
    "svc__C": Real(1e-3, 1e3, prior="log-uniform"),
    "svc__gamma": Real(1e-4, 1e1, prior="log-uniform"),
    "svc__class_weight": Categorical([None, "balanced"]),
}

cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

opt = BayesSearchCV(
    estimator=pipe,
    search_spaces=search_spaces,
    n_iter=80,
    cv=cv,
    scoring="roc_auc",
    n_jobs=-1,
    random_state=42,
    return_train_score=True
)

opt.fit(Xtrain, ytrain)

cv_mean = opt.cv_results_["mean_test_score"]
cv_std  = opt.cv_results_["std_test_score"]
best_idx = np.argmax(cv_mean)

print("\nMejores hiperparámetros encontrados:")
print(opt.best_params_)
print(f"\nMejor AUC promedio (10-Fold): {opt.best_score_:.4f}")
print(f"Desviación estándar del AUC (10-Fold): {cv_std[best_idx]:.4f}")

best_model = opt.best_estimator_
y_pred_all = best_model.predict_proba(X)[:, 1]
auc_total = roc_auc_score(y, y_pred_all)
print(f"AUC total en todo el dataset: {auc_total:.4f}")

RocCurveDisplay.from_estimator(best_model, Xtest, ytest)
plt.title("Curva ROC (SVC RBF · BayesSearchCV)")
plt.tight_layout()
plt.show()
```

  <h3> Modelo 3 - Multi-layer perceptron</h3>
</div>


```python
from sklearn.pipeline import Pipeline # Flujo de trabajo
from sklearn.neural_network import MLPClassifier # Modelo red neuronal
from sklearn.model_selection import StratifiedKFold, cross_val_score # k-folds cross validation
import numpy as np 
from sklearn.gaussian_process import GaussianProcessRegressor # Gaussiano
from sklearn.gaussian_process.kernels import RBF # RBF
```

```python
# Definir columnas
num_columns = [
    'Age', 'Gender', 'Education', 'Country', 'Ethnicity',
    'Nscore', 'Escore', 'Oscore', 'Ascore', 'Cscore',
    'Impulsive', 'SS'
]

cat_columns = []

target = ['Cannabis']

# Separar datos
x = df[num_columns + cat_columns]
y = df[target].values.ravel()

# CV
kfolds = 10
```

```python
# Funcion optimiacion objetivo
def foo(hidden_layers):
    model = MLPClassifier(hidden_layer_sizes=hidden_layers, random_state=10, max_iter=10_000)
    
    pipeline = Pipeline([
        ("scaler", StandardScaler()),
        ("model", model)
    ])

    scores = cross_val_score(pipeline, x, y, scoring="roc_auc", cv=kfolds)
    return np.mean(scores)
```

```python
# Valores iniciales
x1 = [10,5]
x2 = [55,27]
x3 = [100,50]

X = np.vstack([x1,x2,x3])

# Regresion Gaussiana Inicial
y_gp = []
for i in X:
    score = foo(i)
    y_gp.append(score)

kernel = RBF(length_scale=5)
gp = GaussianProcessRegressor(kernel = kernel, random_state=10)
gp.fit(X,y_gp)
```

```python
# Valores aleatorios a probar 
x_capa1 = np.random.uniform(10,100, 1000).reshape(-1, 1)
x_capa2 = np.random.uniform(5, 50, 1000).reshape(-1, 1)
x_gp = np.hstack([x_capa1, x_capa2]).round()

# Optimizacion Bayesiana con 10 iteraciones
for i in range(10):
    gp.fit(X,y_gp)

    y_pred, sigma = gp.predict(x_gp, return_std=True)
    y_upper = y_pred+ 1.96 * sigma
    
    # Elegir el punto score maximo AUC
    max_pred_index = np.argmax(y_upper) 
    max_pred_x_value = x_gp[max_pred_index, :]

    # Evaluar f(x) en el nuevo punto
    y_new = foo(max_pred_x_value.astype(int))

    X = np.vstack([X, max_pred_x_value])
    print(max_pred_x_value)
    y_gp = np.append(y_gp, y_new)
```

```python
# Resultados
best_idx = np.argmax(y_gp)
best_params = X[best_idx]
best_score = y_gp[best_idx]

print(f"Mejor configuración: {best_params}")
print(f"Mejor AUC: {best_score:.4f}")
print(f"AUC promedio: {np.mean(y_gp):.4f}")
print(f"Desviación estándar: {np.std(y_gp):.4f}")
```

Elegimos la metrica roc-auc para evaluar los diferentes modelos.
> - En este caso del consumo de drogas, es común que la proporción de usuarios sea menor que la de no usuarios.
Métricas como la exactitud (accuracy) pueden ser engañosas en este contexto, ya que un modelo que prediga siempre “no consume” podría obtener una precisión alta sin realmente aprender nada útil.
> - El ROC-AUC mide el desempeño del modelo independientemente del umbral de clasificación y es robusto frente al desbalance de clases.
Esto permite comparar modelos sin depender de una proporción específica de positivos/negativos.
> - El AUC representa la probabilidad de que el modelo asigne una puntuación más alta a un usuario real (clase 1) que a un no usuario (clase 0).
En este proyecto, eso significa que el modelo no solo acierta en las etiquetas, sino que distingue correctamente quién tiene mayor riesgo de consumo.
>- Un AUC = 1.0 indica una separación perfecta entre usuarios y no usuarios.
>- Un AUC = 0.5 indica un modelo sin capacidad discriminativa.

---

La regresión logística mostró un rendimiento sólido y estable. El uso de regularización L1 permitió realizar una ligera selección de variables, reduciendo el riesgo de sobreajuste. La diferencia mínima entre el AUC promedio y el AUC total indica buena consistencia del modelo y una adecuada capacidad de generalización.

El modelo SVM fue el que obtuvo el mejor desempeño global, con el AUC más alto tanto en validación cruzada como en el dataset completo. La desviación estándar ligeramente mayor (0.0327) sugiere que su rendimiento varía un poco más entre los pliegues, pero sigue siendo consistente. La inclusión de pesos balanceados ayudó a compensar posibles desbalances en las clases, lo que contribuyó a su excelente capacidad discriminativa.
