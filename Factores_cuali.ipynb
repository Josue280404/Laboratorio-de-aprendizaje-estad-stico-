```python
import pandas as pd
import numpy as np
data = pd.read_excel("aguas_feo.xlsx")
data
```

```python
import statsmodels.api as sm
n = len(data)
X = np.hstack([
    np.ones([n,1]), 
    data['K'].values.reshape([-1, 1]), 
    data['C'].values.reshape([-1, 1]),
    data['H'].values.reshape([-1, 1]),
    data['M'].values.reshape([-1, 1]),
    data['Tamaño'].values.reshape([-1, 1]),
    data['Hielo'].values.reshape([-1, 1])
])
Y = data['Calificación'].values.reshape([-1, 1])

ols = sm.OLS(Y, X)
results = ols.fit()
results.summary()
```
